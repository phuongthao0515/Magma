```python
import json

with open('/home/thaole/thao_le/Magma/datasets/agentnet/agentnet_win_mac_18k.jsonl') as f:
    entry = json.loads(f.readline())

# Top-level fields
print('Top-level fields:')
for k in entry.keys():
    print(f'  {k}')

# Traj step fields
print('\nTraj step fields:')
for k in entry['traj'][0].keys():
    print(f'  {k}')

# Traj step value fields
print('\nTraj step value fields:')
for k in entry['traj'][0]['value'].keys():
    print(f'  {k}')
```

```python
import json
from collections import Counter

apps = Counter()

with open('/home/thaole/thao_le/Magma/datasets/agentnet/meta_data_merged.jsonl') as f:
    for line in f:
        entry = json.loads(line)
        for app in entry.get('applications', []):
            apps[app.lower()] += 1

print(f'Total unique applications: {len(apps)}')
print()
for app, count in apps.most_common():
    print(f'  {app:45s}: {count}')
```
